```python
import pandas as pd
```

```python
df= pd.DataFrame({
    'Id':[1,2,3,4,5,6,7],
    'Name':['John','Meera','David','Geeta','Aka','Tia', 'Rupsi'],
    'Dob':['2011-07-22','2019-01-30','2001-01-2','2016-09-22','2012-12-12','2012-02-22','2011-07-22'],
    'Dob2':['07-2011-22','01-2019-30','01-2001-2','09-2016-22','12-2012-12','02-2012-22','07-2011-22'],
    'Dob3':['22 July, 2011', '22 July, 2019','22 July, 2011','22 July, 2011','22 July, 2011','22 July, 2011','22 July, 2011'],
    'Dob4':['2011','2019','2001','2016','2012','2012','2011']
    
})
df
```

```python
df.dtypes
```

**Strptime()     and      Strftime()**

** 1. Converting Dob and Dob2 in Dtaetime, as they are in different format  **

** 2. Converting Dob3(String Format) [SO STRPTIME()] in Dtaetime, as they are in different format  **

** 3. Converting Dob4(Only Year)  in Proper Dtaetime(d/m/y) **

```python
#Converting Dob and Dob2 in DateTime

df['Dob']= pd.to_datetime(df.Dob)


df['Dob2']= pd.to_datetime(df.Dob, format= '%m-%Y-%d')
```

**Strptime(): Convert String to Datetime Fotmat**

```python
#Converting Dob3, we have to convert it from "Series to DateTime", So Strptime 

from datetime import datetime

def strdte(x):
    return datetime.strptime(x, "%d %B, %Y")


#df['Dob3_New']=list(map(strdte, df.Dob3))
#0R
df['Dob3_New']= df.Dob3.apply(strdte)
```

```python
#Converting Dob4(Only Year), we have to convert it from "Series to DateTime", So Strptime 

from datetime import datetime

def strdte(x):
    return datetime.strptime(x, "%Y")


#df['Dob3']=list(map(strdte, df.Dob3))
#0R
df['Dob4_New']= df.Dob4.apply(strdte)

df
```

**Strftime()  Convert Datetime to String **

#Creating Age Column (Current Date - Dob)

```python
#Craete a column with Todays Date
#(Defualt format is with Time(hh-mm-ss) also, but we don't want Time. We want in YEAR Format so using Strftime)


df['Age']= int(datetime.strftime(datetime.today(), "%Y"))- df.Dob.dt.year.astype(int) 

df 
```

** 2. Extracting  Year, Month, day, weekday from Dob **

```python
df['Year']= df.Dob.dt.year
df['Month']= df.Dob.dt.month
df['Day']= df.Dob.dt.day
df['Weekday_Name']= df.Dob.dt.day_name()
df['Weekday']= df.Dob.dt.weekday

df
```

 ** 3. DatetimeIndex - For Subset**

```python
df[pd.DatetimeIndex(df.Dob).year==2011]
```

```python
df[pd.DatetimeIndex(df.Dob).day_name()=='Friday']
```

**  4. Indexing  **

```python
df.set_index('Dob', inplace=True)

df
```

# 1. Sum(Sales) for a Particular Year or Month

```python
# Summary Function on any Particular Month/Year

df[pd.DatetimeIndex(df.Date).month==1].Sales.max()
```

# 2. Sum(Sales) for a Range of  Year or Month

```python
#For All Dates(overall Data) for 1 Var(Sales)

#This can be done by convering Date as  DateTime and as a Index 

df['Date']= pd.to_datetime(df.Date)

df.set_index('Date', inplace=True)

df.head()
```

```python
df['2009'].Sales.max()
```

```python
#Max(sales) betwee Jan'09 to Dec'12
df['2009-01':'2011-12'].Sales.max()
```

# 3. Sum(Sales) for a Year or Month on Entire Data

```python
#Sum(Sales) group by Year
df['Sales'].resample('Y').sum()
```

```python
#Sum(Sales) group by Month( If we take Month , It automatically takes Year also)
df['Sales'].resample('M').sum()
```

```python
#Sum(Sales) group by Quarter( If we take Qtr , It automatically takes Year also)
df['Sales'].resample('Q').sum()
```

```python
** Sum(sales) group by all Date(Overall Data) can also bt done for specific   PD.GROUPER(Key='',freq='M')
```

```python
# Removing Index as og Now
df.reset_index(inplace=True)
```

# 3. Sum(Sales) for a Year or Month + Any other Var (ie Group by 2 var) on Entire Data

```python
#Sum of 
df.groupby(['Prod_Id', pd.Grouper(key='Date', freq='M')])['Sales'].agg(sum)
```

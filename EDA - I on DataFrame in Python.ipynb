# Rank function in Pandas
--1. Rank based on any variable
--2. Rank with Group by 

**Parameter: ascending=0 means Rank 1 to Heigher value,  method=max(if 2 cobject gat same rank , so it will assign max value)

For Balance we will give Heighest value as Rank 1(so, ascending=0)

```python
df['Bal_Rnk']=df.BALANCE.rank(ascending=0)
```

```python
df[['BALANCE','Bal_Rnk']][df.Bal_Rnk==1]
```

-- Heighest Balance for each category Marital

```python
df['MarBal_Rnk']=df.groupby('MARITAL')['BALANCE'].rank(ascending=0, method='max')
```

```python
df[df.MarBal_Rnk==1]
```

**Checkig if there any duplicate Rank(if there then keep both) , if Yes then create a df**

```python
df[df.Bal_Rnk.duplicated()]
```

```python
df[df.Bal_Rnk.duplicated(keep=False)]
```

**Rank operation on smaller DF for better understanding**

```python
ddf= pd.DataFrame({
                'Id':[1,2,3,4,5,6,7],
                'Marks':[34,22,42,18,42,45,27]
})
```

```python
ddf['Rnk']= ddf.Marks.rank(ascending=0)
ddf
```

-- As 2 and 3rd rank has equal marks, so assigned rank= (2+3)/2  ie 2.5,  If we want rank in whole Number,
   method='max' --> Max rank will assign to both of them , ie 3rd rank(to both) BUT Rnk=2 will be missing
   method='min' --> Min rank will assign to both of them , ie 2rd rank(to both) BUT Rnk=3 will be missing
   method='dense'--> **Best Prctice, No Rank will be misisng** Min Rank ie 2 to both BUT SAME TIME Rnk=3 WILL NOT BE MISSING

```python
ddf['Rnk']= ddf.Marks.rank(ascending=0, method='dense')
ddf
```

# Missing Values in DataFrame

```python
#Craeting a sample DF with missing values
import numpy as np
df= pd.DataFrame({
    'Id':[1,2,3,4,5,6,7],
    'Name':['John','Meera','999999','Geeta','Aka','Tia', np.NaN],
    'Dob':['2011-07-22','2019-01-30','2001-01-2',np.NaN,'2012-12-12','2012-02-22','2011-07-22'],
    'Dob2':[np.NaN,'01-2019-30','01-2001-2','09-2016-22','12-2012-12','02-2012-22',np.NaN],    
    'Sal':[100,200,300,400,np.NaN,600,np.NaN],
    'City':['Mum','Blr','Mum','-8888','Del','Hyd', 'Blr']
})
df
```

```python
#Missing values in DataFrame
df.isna()
```

```python
#Total Missing values in DataFrame COLUMN Wise
df.isna().sum()
```

```python
#Rows having missing values(Eg Row 0, 3 and 6 has saome missing values)
df[df.isna().any(axis=1)]
```

```python
#Total(Sum) of missing values  row wise (Total missing for each Row)
summ= lambda x: x.isna().sum()

df.apply(summ, axis=1)
```

**Replace() : Sometimes missing value is filled with sone dummy numbers, Eg 99999(As 3rd name is 99999)

```python
df= df.replace({
    '99999':'Missing',
    '-8888':'No-City'
})
```

```python
df
```

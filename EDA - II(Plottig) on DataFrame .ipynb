```python
import matplotlib.pyplot as plt
%matplotlib inline
#%matplot... is used just to avoid plot.show()
```

```python
import numpy as np

x= np.arange(0,10)
y= np.arange(11,21)
```

```python
#Plotting matPlot Lib
plt.scatter(x,y, c='g')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter Plot')

#Save Fig.(will get save in Jupyter First page)
plt.savefig('test.png')
```

```python
#Plotting Line with some pattern(--*--)
plt.plot(x,y,'--*')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter Plot')

#Save Fig.
plt.savefig('test.png')


#Other parameter:
#  plt.plot(x,y)
#  plt.plot(x,y,'ro') [For dotted line]
#  plt.plot(x,y, col='r', linestyle='dashed', linewidth=2, markersize=2)
```

```python
# Sub Plots

plt.subplot(2,2,1)
plt.plot(x,y,c='r')
plt.subplot(2,2,2)
plt.plot(x,y,'--*')
plt.subplot(2,2,3)
plt.plot(x,y,'ro')
```

```python
## Bar Plot

x= [9,8,10]
y= [7,11,9]


plt.bar(x,y , align='center')
```

```python
## Histogram  Plot

a=np.array([22,67,45,11,35,60,93,26,77,88])

plt.hist(a)
```

```python
# Box Plot

data=[np.random.normal(0,std,100) for std in range(1,4)]


plt.boxplot(data, vert=True, patch_artist=True)

#If we want plot Horizontally, then Vert=False 
#plt.boxplot(data, vert=False, patch_artist=True)
```

```python
# Pie Plot

labels=['python', 'SAS', 'R', 'SSRS']
share=[45, 35, 18, 2]
col=['green','yellowgreen','blue','lightcoral']
explode= (0.06, 0,0,0)
#If we want to explode all, put values for all

plt.pie(share, explode=explode, labels=labels, colors=col,
autopct='%1.1f%%', shadow=True)
##autopct is for Auto %

plt.axis('equal')
##This for removing errors and warning
```

```python
import seaborn as sns

import pandas as pd
df= pd.read_csv(r'C:\\Users\\ashis\\OneDrive\\Documents\\Ashish\\Python\\EDA\\bank.csv', sep=';')
```

**bank data description:

1 - age (numeric)
2 - job : type of job (categorical: 'admin.','blue-collar','entrepreneur','housemaid','management','retired','self-employed','services','student','technician','unemployed','unknown')
3 - marital : marital status (categorical: 'divorced','married','single','unknown'; note: 'divorced' means divorced or widowed)
:
:
:
19 - euribor3m: euribor 3 month rate - daily indicator (numeric)
20 - nr.employed: number of employees - quarterly indicator (numeric)

Output variable (desired target):
21 - y - has the client subscribed a term deposit? (binary: 'yes','no')

```python
#Correlation of each Variable(Numeric and Float Only)

df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
# Dist plot (uni variate plot) [Only for Numeric/float var]

sns.distplot(df['balance'], kde=True, bins=10)

#kde when we want kd plot ,  bins= wnen we want bins(x asxis will convert into %)
```

```python
# Joint Plot (Bi variate analysis)[Only for Numeric/float var]

sns.jointplot(x='age', y='balance', data=df, kind='reg')
#kind=hex is just to show scatter as hexagon, otherwise dot
#kind=reg, will create reg line(Prob. Density Fun. line) foer each variables
```

** We can see Majority of croud is of balance=0 and Age between 30-40
** Another observation is that there is few OUTLIER( not visible when 'hex' is used)

```python
# Pair plot (multivariate plot) [Only for Numeric/float var]
# It permutate all varibales, Also used to see Correlation between vars

sns.pairplot(df)
```

**In case we want to see all above pair plots but wrt any categorical variables, Eg: for each marital status

```python
sns.pairplot(df, hue='loan')
#Not working for all variables
```

# Plotting Categorical Values

**Count plot (Works both for all vars, but shold be used where count make some sense)
-- My one axis will be Count, so this code will accept only 1 axis(either x or y)

```python
sns.countplot(df.marital)
#or
sns.countplot(x=df.marital)


#If we want Horizontal ploy, use y=
sns.countplot(y=df.marital)
```

```python
##Count plot of marital WRT Loan=(taken / not taken)
sns.countplot(df.marital, hue=df.loan)
```

**Bar plot (Works both for all vars, But out of 2 used vars, ATLEAST 1 SHOULD BE Numeric
-- It Accept only 2 variables for both axis

```python
sns.barplot(x='marital', y='balance', data=df)
```

**Box plot (Works both for all vars, But out of 2 used vars, ATLEAST 1 SHOULD BE Numeric
-- It Accept only 2 variables for both axis

```python
sns.boxplot(x='marital',y='balance', data=df)
```

```python
sns.boxplot(x='marital',y='balance', data=df, hue='loan', orient='v')
```

Married has more outliers 

```python
#Violon plot
#Used to see both Box plot and distribution of data(kde plot)


sns.violinplot(x='balance',y='marital', data=df, hue='loan')
```

Maried and when Loan=No, then there are more outlier
